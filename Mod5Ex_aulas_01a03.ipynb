```python
# 7) seu código aqui 

# Justificativa: Podemos usar a mediana ou a moda para preencher valores ausentes em outras colunas quantitativas ou qualitativas.

def preencher_valores(df, coluna):
    if df[coluna].dtype in ['float64', 'int64']:
        df[coluna].fillna(df[coluna].median(), inplace=True)
    else:
        df[coluna].fillna(df[coluna].mode()[0], inplace=True)

# Aplica a função nas colunas relevantes
for coluna in ['ESCMAE', 'GESTACAO', 'GRAVIDEZ']:
    preencher_valores(sinasc, coluna)

print(sinasc.info())
```

```python
# 8) seu código aqui

sinasc.loc[(sinasc['APGAR5'] >= 8) & (sinasc['APGAR5'] <= 10), 'asfixia_bebe'] = 'normal'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5'] <= 7), 'asfixia_bebe'] = 'asfixia_leve'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5'] <= 5), 'asfixia_bebe'] = 'asfixia_moderada'
sinasc.loc[(sinasc['APGAR5'] >= 0) & (sinasc['APGAR5'] <= 3), 'asfixia_bebe'] = 'asfixia_severa'

frequencias = sinasc['asfixia_bebe'].value_counts()
print(frequencias)
```
